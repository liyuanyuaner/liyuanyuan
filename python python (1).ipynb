```python
"""
@classmethod:è·å–è‡ªèº«ç±»(cls)ä¸­çš„å±æ€§,å¹¶ä¸”å¯ä»¥æ›´æ”¹.
classmethod ä¿®é¥°ç¬¦å¯¹åº”çš„å‡½æ•°ä¸éœ€è¦å®ä¾‹åŒ–ï¼Œä¸éœ€è¦ self å‚æ•°ï¼Œä½†ç¬¬ä¸€ä¸ªå‚æ•°éœ€è¦æ˜¯è¡¨ç¤º
è‡ªèº«ç±»çš„ cls å‚æ•°ï¼Œå¯ä»¥æ¥è°ƒç”¨ç±»çš„å±æ€§ï¼Œç±»çš„æ–¹æ³•ï¼Œå®ä¾‹åŒ–å¯¹è±¡ç­‰

"""
from time import time, localtime, sleep


class Clock(object):
    """æ•°å­—æ—¶é’Ÿ"""

    def __init__(self, hour=0, minute=0, second=0):
        self._hour = hour
        self._minute = minute
        self._second = second

    @classmethod
    def now(cls):
        ctime = localtime(time())
        return cls(ctime.tm_hour, ctime.tm_min, ctime.tm_sec)

    def run(self):
        """èµ°å­—"""
        self._second += 1
        if self._second == 60:
            self._second = 0
            self._minute += 1
            if self._minute == 60:
                self._minute = 0
                self._hour += 1
                if self._hour == 24:
                    self._hour = 0

    def show(self):
        """æ˜¾ç¤ºæ—¶é—´"""
        return '%02d:%02d:%02d' % \
               (self._hour, self._minute, self._second)

def main():
        # é€šè¿‡ç±»æ–¹æ³•åˆ›å»ºå¯¹è±¡å¹¶è·å–ç³»ç»Ÿæ—¶é—´
    clock = Clock.now()
    while True:
        print(clock.show())
        sleep(1)
        clock.run()
        
        
if __name__ == '__main__':
        main()
```

```python
# å¤šè¿›ç¨‹çš„åº“
from multiprocessing import Process
from random import randint
from time import time, sleep
import os


def download_task(filename):
    print('å¼€å§‹ä¸‹è½½%s' % filename)
    time_to_download = randint(5, 10)
    sleep(time_to_download)
    print('%s ä¸‹è½½å®Œæˆ,ç»è¿‡%dç§’' % (filename, time_to_download))


def main():

    start = time()
    # å¼€å¯ä¸¤ä¸ªå¤šè¿›ç¨‹ï¼Œ     å‡½æ•°å              ä¼ é€’çš„å‚æ•°,éœ€è¦æ³¨æ„çš„æ˜¯,å®ƒæ¥å—çš„æ˜¯ä¸€ä¸ªå…ƒç»„(tuple)
    p1 = Process(target=download_task, args=('æ¦‚ç‡è®º.mp4', ))
    p2 = Process(target=download_task, args=('è®ºæ–‡.mp5', ))
    # è·å–è¿›ç¨‹å·

    # å¯åŠ¨è¿›ç¨‹
    p1.start()
    p2.start()

    ##############
    # è¿›ç¨‹é˜»å¡.
    p1.join()
    p2.join()
    ##############
    end = time()
    print(end - start)


if __name__ == '__main__':
    main()

```
